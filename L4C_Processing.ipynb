```r
# Define the paths
extract_folder <- paste0("~/my-public-bucket/GEDI_global_PA_v2/WDPA_GEDI_extract/",iso)
l4c_folder <- paste0("~/my-public-bucket/GEDI_global_PA_v2/WDPA_gedi_L4C_tiles/",iso)
output_folder <- paste0("~/my-public-bucket/GEDI_global_PA_v2/WDPA_GEDI_L2AL2BL4AL4C/",iso) # SPECIFY YOUR OUTPUT FOLDER HERE

# Create output folder if it doesn't exist
if (!dir.exists(output_folder)) {
  dir.create(output_folder, recursive = TRUE)
}

# Function to extract tile number from filename
extract_tile_number <- function(filename) {
  # Look for "Extracted" followed by digits
  if (grepl("Extracted\\d+", filename)) {
    matches <- regmatches(filename, regexpr("Extracted(\\d+)", filename))
    if (length(matches) > 0) {
      tile_num <- as.numeric(gsub("Extracted", "", matches))
      return(tile_num)
    }
  }
  
  # Look for "tile_num_" followed by digits
  if (grepl("tile_num_\\d+", filename)) {
    matches <- regmatches(filename, regexpr("tile_num_(\\d+)", filename))
    if (length(matches) > 0) {
      tile_num <- as.numeric(gsub("tile_num_", "", matches))
      return(tile_num)
    }
  }
  
  return(NA)
}

# Get list of all geopackages in both folders
extract_files <- list.files(extract_folder, pattern = "\\.gpkg$", full.names = TRUE)
l4c_files <- list.files(l4c_folder, pattern = "\\.gpkg$", full.names = TRUE)

# Create data frames with file info
extract_info <- data.frame(
  filepath = extract_files,
  stringsAsFactors = FALSE
)
extract_info$tile_number <- sapply(basename(extract_files), extract_tile_number)
extract_info <- extract_info[!is.na(extract_info$tile_number), ]

l4c_info <- data.frame(
  filepath = l4c_files,
  stringsAsFactors = FALSE
)
l4c_info$tile_number <- sapply(basename(l4c_files), extract_tile_number)
l4c_info <- l4c_info[!is.na(l4c_info$tile_number), ]

# Find matching tile numbers
common_tiles <- intersect(extract_info$tile_number, l4c_info$tile_number)

# Check which tiles have already been processed (NEW CODE)
existing_output_files <- list.files(output_folder, pattern = "combined_tile_\\d+\\.gpkg$", full.names = FALSE)
existing_tile_numbers <- sapply(existing_output_files, function(f) {
  matches <- regmatches(f, regexpr("combined_tile_(\\d+)", f))
  if (length(matches) > 0) {
    return(as.numeric(gsub("combined_tile_", "", gsub("\\.gpkg$", "", matches))))
  }
  return(NA)
})
existing_tile_numbers <- existing_tile_numbers[!is.na(existing_tile_numbers)]

# Filter out tiles that have already been processed
tiles_to_process <- setdiff(common_tiles, existing_tile_numbers)

# Print information about existing files
if (length(existing_tile_numbers) > 0) {
  cat("Found", length(existing_tile_numbers), "existing output files. Skipping tiles:", paste(existing_tile_numbers, collapse = ", "), "\n")
}
if (length(tiles_to_process) > 0) {
  cat("Will process", length(tiles_to_process), "new tiles:", paste(tiles_to_process, collapse = ", "), "\n\n")
} else {
  cat("All tiles have already been processed!\n")
}

# Function to process and save a single tile pair
process_and_save_tile <- function(tile_num) {
  # Get file paths for this tile
  extract_file <- extract_info$filepath[extract_info$tile_number == tile_num]
  l4c_file <- l4c_info$filepath[l4c_info$tile_number == tile_num]
  
  if (length(extract_file) == 0 || length(l4c_file) == 0) {
    return(paste("Error: Could not find files for tile", tile_num))
  }
  
  cat("Processing tile", tile_num, "...\n")
  
  # Read the geopackages
  gedi_extract <- st_read(extract_file[1], quiet = TRUE)
  gedi_l4c <- st_read(l4c_file[1], quiet = TRUE)
  
  # Convert shot_number to character to match data types
  gedi_extract$shot_number <- as.character(gedi_extract$shot_number)
  gedi_l4c$shot_number <- as.character(gedi_l4c$shot_number)
  
  # Combine the data
  combined <- gedi_extract %>%
    left_join(st_drop_geometry(gedi_l4c), by = "shot_number")
  
  # Define output filename
  output_file <- file.path(output_folder, paste0("combined_tile_", tile_num, ".gpkg"))
  
  # Save the combined data
  st_write(combined, output_file, quiet = TRUE)
  
  return(paste("Successfully processed and saved tile", tile_num))
}

# Process only the tiles that haven't been processed yet (MODIFIED)
if (length(tiles_to_process) > 0) {
  results <- sapply(tiles_to_process, process_and_save_tile)
  
  # Print results
  for (result in results) {
    cat(result, "\n")
  }
  
  # Print summary
  cat("\nProcessing complete!\n")
  cat("Processed", length(tiles_to_process), "new tiles\n")
  cat("Skipped", length(existing_tile_numbers), "existing tiles\n")
  cat("Total tiles in output folder:", length(existing_tile_numbers) + length(tiles_to_process), "\n")
  cat("Output files saved to:", output_folder, "\n")
} else {
  cat("No new tiles to process. All matching tiles have already been processed.\n")
  cat("Total existing tiles:", length(existing_tile_numbers), "\n")
  cat("Output folder:", output_folder, "\n")
}
```
